```python
# 3. Load and combine all 18 scRNA-seq reference samples
#
# List of all sample prefixes from your 'tar' command output
sample_prefixes = [
    'GSM6614348_HC-1_', 'GSM6614349_HC-2_', 'GSM6614350_HC-3_',
    'GSM6614351_HC-4_', 'GSM6614352_HC-5_', 'GSM6614353_HC-6_',
    'GSM6614354_UC-1_', 'GSM6614355_UC-2_', 'GSM6614356_UC-3_',
    'GSM6614357_UC-4_', 'GSM6614358_UC-5_', 'GSM6614359_UC-6_',
    'GSM6614360_CD-1_', 'GSM6614361_CD-2_', 'GSM6614362_CD-3_',
    'GSM6614363_CD-4_', 'GSM6614364_CD-5_', 'GSM6614365_CD-6_'
]

# Load each sample and store it in a list
adata_list = []
for prefix in sample_prefixes:
    # Use the 'prefix' argument to specify which sample to load
    adata_sample = sc.read_10x_mtx(scrna_data_path, prefix=prefix, var_names='gene_symbols', cache=False)
    
    # Add the sample ID to the observations
    sample_id = prefix.replace('GSM', '').split('_')[1] # Extracts 'HC-1', 'UC-2', etc.
    adata_sample.obs['sample'] = sample_id
    
    adata_list.append(adata_sample)

# Concatenate all samples into one AnnData object
# The 'join="outer"' keeps all genes from all samples
# The 'label' argument adds the sample ID to the cell barcodes to make them unique
adata_sc = anndata.concat(
    adata_list,
    label="sample",
    keys=[prefix.split('_')[1] for prefix in sample_prefixes],
    join="outer",
    index_unique="-"
)

print("### All 18 scRNA-seq samples loaded and combined:")
print(adata_sc)

# Load the cell annotations
annotations = pd.read_csv(scrna_annot_file, index_col=0)

# Merge annotations with the combined AnnData object
# This assumes the cell barcodes in the annotation file match the barcodes in your data
adata_sc.obs = adata_sc.obs.join(annotations.drop('sample', axis=1))

# Check if merge was successful by looking for non-empty annotation columns
print("\n### Annotation columns after merge:")
print(adata_sc.obs.head())
```
